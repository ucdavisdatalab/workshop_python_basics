## Second, Correct Try

```python
parks = pd.read_csv("parks.csv")
lookup = pd.read_csv("city_state_lookup.csv")
```

```python
lookup.loc[lookup.city == "Arlington", "city"] = "Arlington, Texas"

lookup.loc[lookup.city == "Washington", "city"] = "Washington, D.C."
parks.loc[parks.city == "Washington, DC", "city"] = "Washington, D.C."

lookup.loc[lookup.city == "Saint Paul", "city"] = "St. Paul"

lookup.loc[lookup.city == "Winston–Salem", "city"] = "Winston-Salem"
```

Add St. Paul, Winston-Salem, and Charlotte:

```python
merged = pd.merge(parks, lookup, how = "left", on = "city")
```

```python
merged
```

```python
merged.loc[merged.state.isna()]
```

```python
merged = merged.drop("city_dup", axis = 1)
```

```python
city_state = merged.city.str.split(",")
```

```python
merged["city"] = city_state.str.get(0)
```

```python
state = city_state.str.get(1).str.strip()
```

```python
to_change = ~state.isna() & (state != "D.C.")
```

```python
merged.loc[to_change, "state"] = state[to_change]
```

```python
merged
```

```python
merged.loc[merged.city == "Charlotte/Mecklenburg County", "state"] = "North Carolina"
```

```python
merged.to_csv("parks_final.csv", index = False)
```
