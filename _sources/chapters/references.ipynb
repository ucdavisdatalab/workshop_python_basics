References
==========

This reader provides an introduction to the basics of Python, but there's lots
more to learn. Fortunately, many Python and data science learning resources are
available for free online or through the library.

Here are a few readers and notes created by DataLab staff:

* [README, Write Me!][readme], our workshop about how to organize and document
  computing projects.
* [Principles of Data Visualization][data-viz], our workshop about how to
  design clear and effective data visualizations.
* [Adventures in Data Science][adventures], our course introducing humanities
  undergraduates to data science techniques.
* Nick's [teaching notes][notes] from several years of teaching
  statistical computing.

[readme]: https://ucdavisdatalab.github.io/workshop_how-to-data-documentation/
[data-viz]: https://ucdavisdatalab.github.io/workshop_data_viz_principles/
[adventures]: https://ucdavisdatalab.github.io/adventures_in_data_science/
[notes]: https://github.com/nick-ulle/teaching-notes

And here are a few books created by others that we've found useful:

* [Python for Data Analysis][pydata] (2nd edition) by McKinney. An introduction
  to using Python for data science, by the creator of the Pandas package.
* [Python Data Science Handbook][pyhandbook] by VanderPlas. 

[pydata]: https://wesmckinney.com/pages/book.html
[pyhandbook]: https://jakevdp.github.io/PythonDataScienceHandbook/
